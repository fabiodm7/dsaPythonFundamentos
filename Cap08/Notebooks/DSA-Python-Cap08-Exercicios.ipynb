### Exercício 1

```python
# Crie um array NumPy com 1000000 e uma lista com 1000000.
# Multiplique cada elemento do array e da lista por 2 e calcule o tempo de execução com cada um dos objetos (use %time).
# Qual objeto oferece melhor performance, array NumPy ou lista?

import numpy as np

lista = list(range(1000000))
array = np.arange(1000000)

%time lista2 = [x * 2 for x in lista]
print(lista2[:50])
%time array2 = array * 2
print(array2[:50])
```

### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond. Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])

for x,y,c in zip(xarr,yarr, cond):
    print(x,y,c)

a = [x if c else y for x,y,c in zip(xarr,yarr, cond) ]

print(a)
```

### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas
import sqlite3 as sql
from random import sample

conn = sql.connect('exe10.db')
cur = conn.cursor()
cur.execute('''CREATE TABLE IF NOT EXISTS exercicio
               (pk INTEGER PRIMARY KEY ASC AUTOINCREMENT, aleatorio INTEGER)''')
valores = range(1,100)
lista = sample(valores,k=len(valores))

for i in lista:
    cur.execute('''INSERT INTO exercicio (aleatorio) VALUES (?)''',[i])

conn.commit()

df = pd.read_sql('''SELECT * FROM exercicio''', conn)

conn.close()

df
```
