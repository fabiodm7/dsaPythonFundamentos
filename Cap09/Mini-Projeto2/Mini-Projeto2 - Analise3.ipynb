```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.columns.values
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
sns.set_style("white")
fig, ax = plt.subplots(figsize=(10,5))
cores = ["#0000e6", "#ff1a8c","#a1cc80"]
graf = sns.barplot(data=df,x='fuelType',y='price',hue='gearbox',palette=cores)
graf.yaxis.set_label_text('Média de preço')
graf.xaxis.set_label_text('Tipo de combustível')
graf.set_title('Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio')
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
sns.set_style("white")
fig, ax = plt.subplots(figsize=(10,5))
cores = ["#e60000", "#8cff1a","#80a1cc"]
graf = sns.barplot(data=df,x='vehicleType',y='powerPS',hue='gearbox',palette=cores)
graf.yaxis.set_label_text('Potência média')
graf.xaxis.set_label_text('Tipo de veículo')
graf.set_title('Potência média dos veículos por tipo de veículo e tipo de caixa de câmbio')
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
