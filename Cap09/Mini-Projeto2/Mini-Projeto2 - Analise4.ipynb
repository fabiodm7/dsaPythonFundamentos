```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.columns.values
```

```python
# Calcule a média de preço por marca e por veículo
df2 = df[['brand','vehicleType','price']].groupby(['brand','vehicleType']).mean().astype(int)
df2 = df2.reset_index()
df2
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
df3 = df2.pivot('brand','vehicleType','price')
f, ax = plt.subplots(figsize=(16,24))
sns.heatmap(df3,annot=True,cmap="YlGnBu",ax=ax,linewidths=.1,fmt='.0f')
ax.set_title("Preço médio do veículo por tipo de veículo e marca\n",fontdict= {'size':16})
ax.xaxis.set_label_text('Tipo de veículo')
ax.yaxis.set_label_text('Marca')
```
